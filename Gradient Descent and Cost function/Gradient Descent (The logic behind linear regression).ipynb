### Therefore, the equation of the best fit line (in the form of Y = mX + c) is 

Y = 2X + 3

Gradient Descent is an iterative algorithm to find the optimum values of 'm' and 'c' such that it gives the best fit line for a given set of lines

here, m = 2 and c = 3 from the model

lets find out how did we get them


```python
import numpy as np
```

```python
def gradient_descent(x,y):
    m_curr = c_curr = 0       # starting value for both of them is taken to be 0
    iterations = 10000         # No. of iterations can be fine tuned by trial and error
    learning_rate = 0.01     # can be fine tuned by trial and error
    n = len(x)                # here,in this case, length of x and y is taken to be same. if different , necessary tweaking might be done to remove bugs/errors that may arise
    
    for i in range(iterations):
        y_predicted = m_curr * x + c_curr          # predicting current Y value 
        cost = (1/n) * sum([val**2 for val in (y-y_predicted)])  # setting the cost
#         cost = (1/n) * sum((y - y_predicted)**2) 
        m_derivative = -(2/n) * sum(x * (y - y_predicted))   # partial derivative of m
        c_derivative = -(2/n) * sum(y - y_predicted)         # partial derivative of c
        m_curr = m_curr - learning_rate * m_derivative         
        # adjusting the m_curr value for next iteration and also for defining the size of next baby step 
        c_curr = c_curr - learning_rate * c_derivative  
        # adjusting the c_curr value for next iteration and also for defining the size of next baby step 
        
        print("m {}, c {}, cost {}, iteration {}".format(m_curr,c_curr,cost,i))
        
x = np.array([1,2,3,4,5])
y = np.array([5,7,9,11,13])
# * sum((y - y_predicted)**2)

gradient_descent(x,y)
```

## Results, 
at parameters 




iterations = 10000, learning rate = 0.01 

we get, m = 2.0000000000000218 , c = 2.999999999999924 , ( and also the cost is reducing and minimum at the last iteration among all the iterations)

These values of 'm' and 'c' are almost the same as that of the linear regression model (m = 2 and c = 3). So the global minima has arrived. So we can use these values of 'm' and 'c' to construct the best fitting line. Any query point can be predicted using the line and its equation


but, after iteration 9000, ther is no significant. difference in the (m,c, cost) values of successive iterations. i.e. they become stagant. 

So we can assume it as the threshold point
