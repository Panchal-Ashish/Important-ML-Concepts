### Reading data without specifying the header (i.e. header is default)
so it reads only the 1st row as header even though the data has 2 rows of headers

```python
df = pd.read_excel('2 header data.xlsx')
df

# By default, the header row is taken only as the 1st row. And actual valued dataset is treated henceforth
```

```python
print(df.shape)
```

```python
print(df.info())
```

```python
print(df.describe())
```

```python
print(df.iloc[[0,1,2]])
```

### Reading the data by specifying the no.of headers
So it reads the headers as desired by us

```python
df1 = pd.read_excel('2 header data.xlsx', header = [0,1])
df1

# Here, the header rows are specified. And actual valued dataset is treated henceforth
```

```python
print(df1.shape)
```

```python
print(df1.info())
```

```python
print(df1.describe())
```

```python
print(df1.iloc[[0,1,2]])
```

```python
df1.iloc[[0,1,2],[1,2,4]]
```

## Stacking (with 2 level header file)

level of stacking starts from inside to outside

```python
df1 = pd.read_excel('2 header data.xlsx', header = [0,1])
df1
```

```python
print(f"Stacking with default level (i.e. innermost which is 1(company))\n{df1.stack()}")
print('\n')
print(f"Stacking with level 1 (i.e. innermost which is 1(company)\n{df1.stack(level = 1)}")
# Both give the same result as pandas by default stacks from innermost layer

print('\n')
print(f"Stacking with level 0 (i.e. with price and P/E ratio) \n{df1.stack(level = 0)}")
```

```python
df1_stacked = df1.stack()
df1_stacked
```

## Unstacking (for the 2 level stacked file)

```python
df1_stacked.unstack(level = 1)
```

## Stacking (for 3 level header file)

```python
df2 = pd.read_excel('3 header data.xlsx', header = [0,1,2])
df2
```

```python
df2_stacked_2 = df2.stack()  # here the innermost level is 2, so default level is taken as 2
df2_stacked_2 
```

```python
df2_stacked_1 = df2.stack(level = 1)
df2_stacked_1
```

## Unstacking (for 3 level stacked file)

```python
unstacked_3_level_level2 = df2_stacked_2.unstack()
unstacked_3_level_level2 
```

```python
unstacked_3_level_level2.to_csv("unstacked_3_level_level2.csv",index = False)
```
